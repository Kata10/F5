## Filtrado de las variables "Country" y "Customer spendings" para luego graficar los resultados

```python
# Primero agrupamos los datos de la variable "Country". 
# Luego agregamos la variable "Customer spendings" y le indicamos que sume los valores.
# Al final le asignamos un nombre a la variable (en este caso grouped) para hacer operaciones sobre la agrupación 



grouped = df.groupby("Country").agg({
    "Customer spendings": 'sum',
    
})

grouped[grouped["Customer spendings"] > 10]
```

```python
#Para graficar las variables obtenidas, pimero importamos matplotlib.
#luego le indicamos el tipo de gráfico deseados:(kind = bar) 

import matplotlib.pyplot as plt



grouped["Customer spendings"].plot(kind = "bar")

plt.show()
```
